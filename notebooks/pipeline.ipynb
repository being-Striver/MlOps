```python
from sklearn.pipeline import Pipeline

#feature scaling

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
#define steps
steps=[("standard_scalar",StandardScaler()),
       ("classifier", LogisticRegression())]
```

```python
steps
```

```python
pipe=Pipeline(steps=steps)
```

```python
#we have technique to visualize the pipeline in efficient way.
from sklearn import set_config
set_config(display='diagram')
```

```python
pipe
```

```python
#creating a dataset
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(
    X, y,
    test_size=0.33, random_state=42
)
```

```python
X_train
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
#complex example of columns transformers
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
numeric_processor= Pipeline(
    steps=[
        ("impute_mean", SimpleImputer(missing_values=np.nan,strategy='mean' )),
        ('scalar', StandardScaler())
    ]
)
```

```python
numeric_processor
```

```python
#categorical processing pipeline

from sklearn.preprocessing import OneHotEncoder
category_processor= Pipeline(
    steps=[
        ("impute_constant", SimpleImputer(fill_value="missing", strategy="constant" )),
        ('ohe', OneHotEncoder(handle_unknown="ignore"))
    ]
)
```

```python
category_processor
```

```python
#combine both pipeline
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer(
    [
        ("categorical", category_processor, ["gender", "city"]),
        ("numerical", numeric_processor, ["age", "height"])
    ]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
model_pipe=make_pipeline(preprocessor, LogisticRegression())
```
